```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16], 'ro')
plt.axis([0,6,0,10])
plt.show()
```

```python
t = np.arange(0,5,0.2)
plt.plot(t, t ,'r--', t , t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
area = np.random.rand(50) * 100
plt.scatter(x,y,s=area, c=colors, alpha=0.5)
```

```python
objects = ('a', 'b', 'c', 'd', 'e', 'f')
y_pos = np.arange(len(objects))
value = [1, 2, 3, 4, 5, 6]
plt.bar(y_pos, value, alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Title')
plt.show()
```

```python
labels = ('Python', 'C', 'Ruby', 'Java')
sizes = (10,20,30,40)
colors = ['red', 'green', 'yellow', 'blue']

plt.pie(sizes,labels=labels, colors=colors,autopct='%1.0f%%',shadow=True)
plt.axis('equal')

plt.show()
```

```python
x = np.arange(5)
y = [2,4,6,8,10]
y2 = [10,11,12,13,14]

plt.plot(x, y, 'r', label='Y')
plt.plot(x, y2,'b',label='Y2')
# plt.legend()
plt.show()
```
